```python
import requests
from bs4 import BeautifulSoup
import time

# 讓使用者輸入月份
target_month = input("請輸入想要查詢的月份 (1-12): ").strip()
url = "https://www.ptt.cc/bbs/Gossiping/index.html"
headers = {
    'cookie': 'over18=1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

print(f"\n--- 正在搜尋 {target_month} 月文章，請耐心等候翻頁 ---")

# 設定較大的翻頁上限，確保能爬到 4 月份
for i in range(1500):
    try:
        res = requests.get(url, headers=headers, timeout=10)
        if res.status_code != 200:
            time.sleep(2)
            continue
            
        soup = BeautifulSoup(res.text, 'html.parser')
        articles = soup.find_all('div', class_='r-ent')
        
        # 逐一檢查該頁文章
        for art in reversed(articles):
            date_div = art.find('div', class_='date')
            if not date_div: continue
            
            date_raw = date_div.text.strip()
            month = date_raw.split('/')[0]
            
            # 符合目標月份才顯示
            if month == target_month:
                if art.find('a'): # 確保文章未被刪除
                    title = art.find('a').text
                    if "[公告]" in title: continue # 依作業需求過濾掉公告
                    
                    author = art.find('div', class_='author').text
                    link = "https://www.ptt.cc" + art.find('a')['href']
                    
                    # 輸出作業要求的四項資訊
                    print(f"日期：{date_raw}")
                    print(f"作者：{author}")
                    print(f"標題：{title}")
                    print(f"連結：{link}")
                    print("-" * 30)

        # 檢查是否已經翻過頭 (到前一個月份了)
        first_date = articles[0].find('div', class_='date').text.strip()
        first_month = int(first_date.split('/')[0])
        
        # 如果爬到比目標月份還舊的月份，就停止
        if first_month < int(target_month) and first_month != 0:
            # 排除 1 月跨 12 月的情況
            if not (first_month == 1 and target_month == "12"):
                break
            
        # 繼續翻往上一頁
        prev_btn = soup.find('a', string='‹ 上頁')
        if prev_btn:
            url = "https://www.ptt.cc" + prev_btn['href']
            # 每翻 20 頁印出一個小提示，讓你知道程式還在跑
            if i % 20 == 0:
                print(f"(正在翻頁搜尋中... 目前第 {i} 頁)")
        else:
            break
            
    except Exception as e:
        continue

print("\n--- 搜尋任務完成！ ---")
```
